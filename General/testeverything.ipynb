```python
import pandas as pd
d = {'col1':[1,2,3,4],'col2':[1,'ew1','ew2','ew3'],'col3':[5,3,2,1],'col4':[2,'re','re','ww']}
df = pd.DataFrame(d)
df
```

```python
# df.iloc[:,1].astype(str)
# type(df.iloc[0,1])
df.iloc[:,1] = df.iloc[:,1].apply(lambda col: str(col))
```

```python
isinstance(df.iloc[:,1][0],str)
```

```python
# 如果是 str 則用 Label encoding 轉換
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
le = preprocessing.LabelEncoder()
for i in range(df.shape[1]):
    for j in range(df.shape[0]):
        if isinstance(df.iloc[:,i][j],str):
            df.iloc[:,i] = df.iloc[:,i].apply(lambda col: str(col))
            df.iloc[:,i] = le.fit_transform(df.iloc[:,i])
            # print(df.iloc[:,i][j])
            break
```
