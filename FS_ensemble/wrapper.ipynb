## 切成5個驗證檔案 5-fold file


```python
temp = pd.read_excel('/Users/emily/Desktop/Research/oversampling_python/data/bank/bankrupt.xlsx')
```

```python
origin = temp
last = origin.shape[1]-1
origin['Bankrupt?'] = origin['Bankrupt?'].map({1:'positive',0:'negative'})
```

```python
i = pd.Index([temp.columns[-1]])
```

```python
from sklearn.model_selection import KFold
import numpy as np
floder = KFold(n_splits=5, random_state=42, shuffle=True)
train_files = []   # 存放5折的训练集划分
test_files = []     # # 存放5折的测试集集划分
for k, (Trindex, Tsindex) in enumerate(floder.split(temp)):
        train_files.append(np.array(temp)[Trindex].tolist())
        test_files.append(np.array(temp)[Tsindex].tolist())

    # 把划分写入csv,检验每次是否相同
df = pd.DataFrame(data=train_files, index=['0', '1', '2', '3', '4'])
# with open('train_patch.csv','w') as file:
    
#      writer = csv.writer(file)
df.to_excel(r'../data/bank/train_patch.xlsx')
df1 = pd.DataFrame(data=test_files, index=['0', '1', '2', '3', '4'])
df1.to_excel(r'../data/bank/test_patch.xlsx')
```

```python
for i in range(5):
    trainframe = pd.DataFrame(train_files[i],columns=temp.columns)
    testframe = pd.DataFrame(test_files[i],columns=temp.columns)
    trainframe.to_excel(r'../data/bank/bank-tra'+str(i)+'.xlsx')
    testframe.to_excel(r'../data/bank/bank-tst'+str(i)+'.xlsx')
```
