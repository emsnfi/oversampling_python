```python
import os

filepath = '/Users/emily/Desktop/Research/oversampling_python/data/NSL-KDD-1/KDDTrain+.arff'
file = open(filepath)
lines = file.readlines()
content = []
columns = []
datarea = False
for l in lines:
    if 'attribute' in l:
        temp = l.split(' ')
        temp[1] = eval(temp[1])
        columns.append(temp[1])
    else:
        if 'data' in l:
            datarea = True
            continue
        if datarea:
            ctemp = l.split(',')
            content.append(ctemp)
        
```

```python
import pandas as pd
df = pd.DataFrame(content,columns=columns,index=None)
```
